```python
# import package
import numpy as np
import pandas as pd

# import library untuk visualisasi
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
# data Keseluruhan
data = pd.read_csv('car_dekho.csv')

print("DATA KESELURUHAN")

display(data.head(20))
```

```python
# ukuran data
data.shape

# info data
data.info()

# deskripsi data
data.describe()
```

```python
# Scatter plot with day against tip
plt.scatter(data['year'], data['selling_price'], c=data['km_driven'])
 
# Adding Title to the Plot
plt.title("Scatter Plot - Car Dekho")
 
# Setting the X and Y labels
plt.xlabel('year')
plt.ylabel('selling_price')

plt.colorbar()
plt.show()
```

```python
# operasi visualisasi
sns.pairplot(data, x_vars=['selling_price', 'km_driven','fuel'], 
             y_vars='year', size=4, aspect=1, kind='scatter')
plt.show()
```

```python
sns.heatmap(data.corr(), cmap="YlGnBu", annot = True)
plt.show()
```

```python
# Scatter plot with day against tip
plt.plot(data['year'],data['selling_price'])
plt.plot(data['year'],data['km_driven'])
 
# Adding Title to the Plot
plt.title("Scatter Plot - Car Dekho")
 
# Setting the X and Y labels
plt.xlabel('year')
plt.ylabel('selling_price')
 
plt.show()
```
